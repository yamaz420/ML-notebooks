```python
x = np.linspace(-4,8,num=100)
```

```python
fx = []
for i in range(len(x)):
    fx.append(x[i]**2)
```

```python
fx2 = []
# y = 0.6*x^2 -2 x +5 
for i in range(len(x)):
    fx2.append(0.6*x[i]**2 + -2*x[i]  + 5)
```

```python
fx3 = []
# y = 1.5*x^2 -2 x -10 
for i in range(len(x)):
    fx3.append(1.5*x[i]**2 + -2*x[i]  - 10)
```

```python
plt.plot(x,fx)
plt.plot(x,fx2)
plt.plot(x,fx3)
plt.grid()
#plt.axvline()
#plt.axhline()
plt.show()
```
